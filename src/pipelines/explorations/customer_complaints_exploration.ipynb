## Explore Silver Layer Transformations

```python
%%sql
-- Check age categorization
SELECT 
  age_category,
  COUNT(*) as count,
  AVG(days_open) as avg_days_open
FROM lakeflow_dec_pipe_r_scripts.silver_complaints_categorized
GROUP BY age_category
ORDER BY avg_days_open;
```
